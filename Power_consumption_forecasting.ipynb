IMPUTING DATA :

```python
import numpy as np
# fill missing values with a value at the same time one day ago
def fill_missing(values):
    one_day = 60 * 24
    for row in range(values.shape[0]):
        for col in range(values.shape[1]):
            if np.isnan(values[row, col]):
                values[row, col] = values[row - one_day, col]
# fill missing
fill_missing(df.values)                
```

```python
# add a column for for the remainder of sub metering
values = df.values
df['sub_metering_4'] = (values[:,0] * 1000 / 60) - (values[:,4] + values[:,5] + values[:,6])
# save updated dataset
df.to_csv('household_power_consumption.csv')
```

```python
# resample data to monthly
month_data = dataset.resample('M').mean()
month_data.shape
X_month=month_data[['Global_reactive_power','Voltage','Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']]
y_month=month_data['Global_active_power']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(X_month);
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_month, y_month, test_size=0.5,  random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
y_pred = linreg.predict(X_test)
from sklearn.metrics import r2_score
print(r2_score(y_test , y_pred))
print('Testing score: ',linreg.score(X_test,y_test))
print('Training Score: ',linreg.score(X_train,y_train))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
rf = RandomForestRegressor(max_depth=2, random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))
print('Training Score: ', rf.score(X_train, y_train))
print('Testing score: ', rf.score(X_test, y_test))
```

```python
from sklearn.ensemble import AdaBoostRegressor
adb = AdaBoostRegressor(rf, n_estimators=300)
adb.fit(X_train, y_train)
y_pred = adb.predict(X_test)
print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))
print('Training Score: ', adb.score(X_train, y_train))
print('Testing score: ', adb.score(X_test, y_test))
```
